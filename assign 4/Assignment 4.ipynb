```python
x=df.drop(['Price'],axis=1)
y=df['Price']
y1=y.to_numpy().reshape(-1,1)
x,y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
kfold = KFold(5, True, 1)
beta=[]
r2=[]
intercept=[]
best_intercept=[]
best_beta=[]
best=0
for train_index, val_index in kfold.split(x,y): 
    X_train, X_test = x[train_index], x[val_index] 
    y_train, y_test = y[train_index], y[val_index]
    mean_y=np.mean(y_test)
    model=LinearRegression()
    model.fit(X_train,y_train)
    beta.append(model.coef_)
    intercept.append(model.intercept_)
    y_pred=model.predict(X_test)
    print(y_pred)
    ss_tot= sum((y_test - mean_y) ** 2)
    ss_res= sum((y_test - y_pred) ** 2)
    c = 1 - (ss_res/ss_tot)
    if best<c:
      best=c
      best_beta=model.coef_
      best_intercept=model.intercept_
    r2.append(c)
print('r2 scores:',r2)
print('beta matrix:',beta)
print('best beta value:',best_beta)
print('best intercept value',best_intercept)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
x_train ,x_test ,y_train,y_test= train_test_split(x,y1,test_size=0.3)
best_beta=best_beta.reshape(-1,1)
y2=np.matrix(x_test)*np.matrix(best_beta)+best_intercept
print(y_test,y2)
print(r2_score(y_test,y2))
```

```python
df=pd.read_csv('USA_Housing.csv')
df.head()
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
y1=y.to_numpy().reshape(-1,1)
x,y1
```

```python
from sklearn.model_selection import train_test_split

x2, x_test, y2, y_test = train_test_split(x,y1,test_size=0.3)
x_train, x_val, y_train, y_val = train_test_split(x2,y2,train_size =0.7)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_train=np.insert(x_train, 0, 1, axis=1)
y=y_train.reshape(-1,1)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
  m=len(X)
  for i in range(iterations):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sum_delta = (alpha / m) * X.transpose().dot(errors);
    theta = theta - sum_delta;  
  return theta
```

```python
iterations = 1000;
alpha = [0.001,0.01,0.1,1]
m=len(x_train)
coef=[]
for i in alpha:
  theta = np.zeros(6).reshape(-1,1)
  theta= gradient_descent(x_train, y, theta, i, iterations)
  theta=theta.reshape(1,-1)
  coef.append(theta[0].tolist())
print(coef)
```

```python
from sklearn.metrics import r2_score
scaler = MinMaxScaler()
x_val = scaler.fit_transform(x_val)
x_val=np.insert(x_val, 0, 1, axis=1)
x_test = scaler.fit_transform(x_test)
x_test=np.insert(x_test, 0, 1, axis=1)
res_r2_val=[]
res_r2_test=[]
for th in coef:
  my_array = np.array(th)
  y_pr=x_val.dot(my_array)
  y_pr=y_pr.reshape(-1,1)
  res_r2_val.append(r2_score(y_val,y_pr))
for th in coef:
  my_array = np.array(th)
  y_pr=x_test.dot(my_array)
  y_pr=y_pr.reshape(-1,1)
  res_r2_test.append(r2_score(y_test,y_pr))
print(res_r2_val)
print(res_r2_test)
idx=res_r2_test.index(max(res_r2_test))
best_coeff=coef[idx]
print(best_coeff)
```
